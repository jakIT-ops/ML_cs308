```python
import numpy as np
from math import log2
```

```python
# Бодит утга
y_true = [1, 1, 2, 2, 4]  # y_true  y(original values)
# Таамагласан утга
y_pred = [0.6, 1.29, 1.99, 2.69, 3.4]  # y_pred = y
```

Mean square error: (pred - true)^ 2

```python
# Mean square error - MSE - дундуаж квадратын алдаа

MSE = np.square(np.subtract(y_true, y_pred)).mean()  # mean() арифметик дундаж олно
print('MSE', MSE)
```

```python
def MSE(y_true, y_pred):
  sum_square_error = 0.0
  for i in range(len(y_true)):
    sum_square_error += (y_true[i] - y_pred[i]) ** 2.0
  mean_square_error = 1.0 / len(y_true) * sum_square_error 
  return mean_square_error
print(MSE(y_true, y_pred))
```

```python
# Mean Absolute Error
def MAE(y_true, y_pred):
  sum_abs_error = 0.0
  for i in range(len(y_true)):
    abs_error = np.abs(y_pred[i] - y_true[i])
    sum_abs_error += abs_error
  mae = sum_abs_error / len(y_true)
  return mae
print(MAE(y_true, y_pred))
```

```python
# Mean Bias Error - MBE
# Дундаж хэвийсэн алдааг үндсэндээ загварын дундаж хэвийх утгыг тооцоолох,
# загварын хазайлтыг засахын тулд ямар нэгэн алхам хийх шаардлагатай эсэхийг шийдэхэд ашигладаг.
def MBE(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_true = y_true.reshape(len(y_true), 1)  # reshape() массивын хэлбэрийн өөрчлөх
    y_pred = y_pred.reshape(len(y_pred), 1)
    diff = (y_true - y_pred)
    MBE = diff.mean()  # mean() арифметик дундаж олно
    print('MBE = ', MBE)
MBE(y_true, y_pred)
```

```python
# Hinge Loss/Multi class SVM Loss
# Гүдгэр функц
# Ангилал хийх үнэлгээг гаргана.
# Тухайлбал зөв ангилалд орж байгаа өгөгдлийн үнэлгээ буруугаасаа өндөр байх ёстой
#  Hinge loss нь ангиллын алдагдлын функц бөгөөд SVM алгоритмын гол хэсэг юм
import numpy as np

def hingeforsingle(feature, theta, label):
    y = np.matmul(theta/10, feature)
    hingeloss = np.max([0.0, (1 - label*y)])
    return hingeloss

feature = np.array([2,4,4,3,6,9,7,4])
theta = np.array([3,3,3,3,-3,-3,-3,-3])

label = 1

hingeloss = hingeforsingle(feature, theta, label)
print(hingeloss)
```

Cross entropy : Classify ангилддаг загварт түлхүү ашигладаг 

```python
# Cross Entropy Loss Function
# Өгөгдөл нь нэг анги эсвэл өөр ангилалд хамаарах магадлалыг таамаглах замаар өгөдлийг ангилдаг
# машин сургалтын ангилалын загваруудыг сургахад ашигладаг оновчлолын функц юм.

def cross_entropy(p, q):
    return - sum([p[i]*log2(q[i]) for i in range(len(p))])

q = [0.10, 0.40, 0.50]
p = [0.80, 0.15, 0.05]
ce_pq = cross_entropy(p, q)
# calculate cross entropy H(P, Q)
print('H(P, Q): %.3f bits' % ce_pq)
# calculate cross entropy H(Q, P)
print('H(Q, P): %.3f bits' % ce_pq)
```

```python
def binary_cross_entropy(y_true, y_pred):
  sum_score = 0.0
  for i in range(len(y_true)):
    sum_score += y_true[i] * log2(1e-15 + y_pred[i])
  mean_sum_score = 1.0 / len(y_true) * sum_score
  return -mean_sum_score
print(binary_cross_entropy(p,q))
```
